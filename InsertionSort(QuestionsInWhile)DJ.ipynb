```python
import numpy as np
import matplotlib.pyplot as plt
npermut = 0
total = 0
cases = np.zeros(1001)
p=[]
def isortSteps(a):
    #print (a)
    v = []
    for i in range(len(a)):
        v.append(a[i])
        
    steps = 0
    for i in range(1,len(v)):
        x = v[i]
        j = i-1
        steps=steps+1
        while (j > -1) and (v[j] > x):
            v[j+1] = v[j]
            j = j -1
            steps=steps+1
        v[j+1] = x
    #print (steps)
    return steps

def swap(v,i,j):
    t = v[i]
    v[i] = v[j]
    v[j] = t

def perm(v,i,n):
    global npermut
    global total
    global cases
    
    j=0
    k=0
    l=0
    if i == n:
        npermut = npermut + 1
        k = isortSteps(v)
        p.append(k)
        total = total + k
        cases[k] = cases[k] + 1
    else:
        for j in range (i,n):
            swap (v, i, j)
            perm (v, i+1, n)
            swap (v, i, j)
            
def fact(n):
    f = n
    ax = 1
    while f > 0:
        ax = ax*f
        f = f-1
    return ax            

n=4
v=[]
for i in range(n):
    v.append(i+1)

factorial = fact(n)    
perm(v,0,len(v))

promedio = 0
suma = 0    
for h in range (0,factorial):
    suma = suma + p[h]
    
promedio = (suma)/factorial
print ("el promedio es : " , promedio)
print (p)
plt.hist(p,factorial,(min(p),max(p)))
plt.xlabel('Steps')
plt.ylabel('Frequency')
plt.legend(['Steps While on Insertion Sort']);
```
